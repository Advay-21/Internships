```python
# DISCLAIMER: 'df' refers to the data you passed in when calling 'dtale.show'

import pandas as pd

if isinstance(df, (pd.DatetimeIndex, pd.MultiIndex)):
	df = df.to_frame(index=False)

# remove any pre-existing indices for ease of use in the D-Tale code, but this is not required
df = df.reset_index().drop('index', axis=1, errors='ignore')
df.columns = [str(c) for c in df.columns]  # update columns to strings in case they are numbers

chart_data = pd.concat([
	pd.Series(df.set_index('SETTLEMENTDATE').index.to_period('W').to_timestamp(how='end').values, index=df.index, name='SETTLEMENTDATE||W'),
	df['TOTALDEMAND'],
], axis=1)
chart_data = chart_data.sort_values(['SETTLEMENTDATE||W'])
chart_data = chart_data.rename(columns={'SETTLEMENTDATE||W': 'x'})
chart_data_max = chart_data.groupby(['x'], dropna=True)[['TOTALDEMAND']].max()
chart_data_max.columns = ['TOTALDEMAND||max']
chart_data = chart_data_max.reset_index()
chart_data = chart_data.dropna()

import plotly.graph_objs as go

charts = []
line_cfg = {'line': {'shape': 'spline', 'smoothing': 0.3}, 'mode': 'lines'}
charts.append(go.Scatter(
	x=chart_data['x'], y=chart_data['TOTALDEMAND||max'], name='TOTALDEMAND||max', **line_cfg
))
figure = go.Figure(data=charts, layout=go.Layout({
    'legend': {'orientation': 'h', 'y': -0.3},
    'title': {'text': 'Maximum of TOTALDEMAND by SETTLEMENTDATE (Weekly)'},
    'xaxis': {'title': {'text': 'SETTLEMENTDATE (Weekly)'}},
    'yaxis': {'title': {'text': 'Maximum of TOTALDEMAND'}, 'type': 'linear'}
}))

# If you're having trouble viewing your chart in your notebook try passing your 'chart' into this snippet:
#
from plotly.offline import iplot, init_notebook_mode

init_notebook_mode(connected=True)
for chart in charts:
    chart.pop('id', None) # for some reason iplot does not like 'id'
iplot(figure)
```

```python
def rmse(predict,actual):
    return np.sqrt(((predict-actual)**2).mean())
```
